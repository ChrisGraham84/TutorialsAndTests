```python
for word in ['Call','me','Ishmael']:
    print(word)
```

```python
sent1 = ['Call', 'me', 'Imhmael', '.']
for xyzzy in sent1:
    if xyzzy.endswith('l'):
        print(xyzzy)
```

```python
for token in sent1:
    if token.islower():
        print(token, 'is a lowercase word')
    elif token.istitle():
        print(token, 'is titlecase word')
    else:
        print(token, 'is puncutation')
```
